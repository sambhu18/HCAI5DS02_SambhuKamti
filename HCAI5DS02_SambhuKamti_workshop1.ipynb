# 1. Sports Performance:

```python
import numpy as np
player_heights = [5.1, 4.8, 5.0, 4.9, 5.3, 4.7, 4.9, 4.8, 5.2, 5.0]

# mean
mean_height = np.mean(player_heights)
print(f"Mean Height: {mean_height:} ft")

# Standard deviation
std_height = np.std(player_heights, ddof=1)
print(f"Standard Deviation: {std_height:} ft")

# Coefficient of variation
cv_height = (std_height / mean_height) * 100
print(f"Coefficient of Variation: {cv_height:}%")

```

–Are the player’s times tightly clustered or highly variable?
=> The coefficient for variation is 3.799 which means the player's times are little seperated so this is tightly clustered.

# – If one player was later found to have mistakenly reported 3.8 seconds, recalculate and
# explain the impact of outliers.

```python
heights_with_outlier = [5.1, 4.8, 5.0, 4.9, 5.3, 4.7, 4.9, 4.8, 5.2, 3.8]

mean_outlier = np.mean(heights_with_outlier)
std_outlier = np.std(heights_with_outlier, ddof=1)
cv_outlier = (std_outlier / mean_outlier) * 100

print(f"New Mean: {mean_outlier:} ft")
print(f"New Std Dev: {std_outlier:} ft")
print(f"New CV: {cv_outlier:}%")

# => If one player was later found to have mistakenly reported 3.8 seconds then the impact of the outlier is that the mean decreses which causes increse in standard deviation (data are more spread out from the mean) so because of that the data ar less tightly clustered.
```

2. Patient Blood Pressure Readings:
Task 3.2

```python
bp = [118, 122, 125, 130, 135, 138, 142, 144, 146, 150, 152, 155, 160, 162, 165]

#– Compute the mean, median, and standard deviation.
mean_bp = np.mean(bp)
median_bp = np.median(bp)
std_bp = np.std(bp, ddof=1)

print(f"Mean: {mean_bp:}")
print(f"Median: {median_bp}")
print(f"Standard Deviation: {std_bp:}")
```

#– Plot a histogram. Is the distribution skewed?
# => The histogram have higher values stretching out which means it is slightly right-skewed.

```python
import matplotlib.pyplot as plt

plt.hist(bp, bins=6, color='skyblue', edgecolor='black')
plt.title("Systolic Blood Pressure")
plt.xlabel("BP (mmHg)")
plt.ylabel("Frequency")
plt.show()
```

#– Compute the IQR. Identify any patients with unusually high blood pressure


```python
q1 = np.percentile(bp, 25)
q3 = np.percentile(bp, 75)
iqr = q3 - q1

print(f"IQR: {iqr}")
# Outlier rule
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr
print(f"Outliers above {upper_fence}")
```

– Explain whether these statistics support that the clinic population has a normal range of
BP levels.
=> This showed that the mean ~144, SD ~14 and it is slightly skewed which is perfectly normal where some patients unusually high.

3. Retail Sales Summary:
Task 3.3

```python
sales = [212, 198, 245, 210, 230, 185, 270, 205, 190, 250, 260, 225, 215, 195]

#– Calculate: Mean, median, mode, range, variance, standard deviation
from scipy import stats

mean_sales = np.mean(sales)
median_sales = np.median(sales)
mode_sales = stats.mode(sales, keepdims=True)[0][0]
range_sales = np.ptp(sales)
var_sales = np.var(sales, ddof=1)
sd_sales = np.std(sales, ddof=1)

print(f"Mean: {mean_sales}")
print(f"Median: {median_sales}")
print(f"Mode: {mode_sales}")
print(f"Range: {range_sales}")
print(f"Variance: {var_sales}")
print(f"Std Dev: {sd_sales}")
```

# – Draw a bar chart of daily sales and annotate any highs/lows


```python
plt.bar(range(len(sales)), sales, color='orange')
plt.title("Daily Sales")
plt.ylabel("Sales")
plt.xlabel("Day")
plt.axhline(mean_sales, color='red', linestyle='--', label=f"Mean = {mean_sales:}")
plt.legend()
plt.show()
```

```python
# Case 5 - E - Commerce Performance:

#– Calculate and compare CVs
import numpy as np
np.random.seed(42)
returning_users = np.random.normal(120, 15, 30)
new_users = np.random.normal(90, 25, 30)
mean_returning = np.mean(returning_users)
std_returning = np.std(returning_users, ddof=1)
mean_new = np.mean(new_users)
std_new = np.std(new_users, ddof=1)
cv_returning = (std_returning / mean_returning) * 100
cv_new = (std_new / mean_new) * 100
print(f"Returning Users - Mean: ${mean_returning:.2f}, CV: {cv_returning:.2f}%")
print(f"New Users - Mean: ${mean_new:.2f}, CV: {cv_new:.2f}%")
if cv_returning > cv_new:
    print("Returning users are more variable relative to their mean.")
else:
    print("New users are more variable relative to their mean.")


```

```python
#Case 1 - Daily Sales Trends in Two Product Categories:

import pandas as pd
df = pd.read_csv(r"C:\Users\Dell\Desktop\HCAI5DS02_SauravKafle\venv\daily_sales.csv")

# • Tasks:
# – Visualize the distribution of total sales for each category using boxplots and histograms.Interpret central tendency and spread.
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='category', y='total_sales', data=df)
plt.title('Boxplot of Total Sales by Category')
plt.show()

# Histogram
plt.figure(figsize=(10,6))
sns.histplot(data=df[df['category'] == 'A'], x='total_sales', color='blue', label='Category A', kde=True, alpha=0.6)
sns.histplot(data=df[df['category'] == 'B'], x='total_sales', color='orange', label='Category B', kde=True, alpha=0.6)
plt.title('Histogram of Total Sales by Category')
plt.legend()
plt.show()

# Interpretation:
# =>The central tendency is shown by where the bulk of the data is (median in boxplot, peaks in histogram).The spread is the width of the boxes and how spread out the histograms are. Wider means more variable.
```

```python
#– Create a time series plot of daily total sales. Are there visible trends or outliers?
print(df.columns)
plt.figure(figsize=(12,6))
sns.lineplot(data=df[df['category']=='A'], x='date', y='total_sales', label='Category A')
sns.lineplot(data=df[df['category']=='B'], x='date', y='total_sales', label='Category B')
plt.xticks(rotation=45)
plt.title('Daily Total Sales Over Time')
plt.legend()
plt.show()
# There are visible Increasing/Decreasing trends
```

```python
#– Calculate Coefficient of Variation (CV) for total sales and avg basket for each category.Which category is more variable?
cv_data = []

for cat in ['A', 'B']:
    df_cat = df[df['category'] == cat]
    for col in ['total_sales', 'avg_basket']:
        mean_val = df_cat[col].mean()
        std_val = df_cat[col].std()
        cv = (std_val / mean_val) * 100
        cv_data.append((cat, col, cv))

import pandas as pd
cv_df = pd.DataFrame(cv_data, columns=['Category', 'Metric', 'CV'])
print(cv_df)
# Higher CV category is more relative variability.
```

```python
#– Compute and visualize a 7-day moving average for both categories. Discuss stability and implications for business planning

df['date'] = pd.to_datetime(df['date'])

df = df.sort_values('date')
df['total_sales_ma'] = df.groupby('category')['total_sales'].transform(lambda x: x.rolling(7, min_periods=1).mean())

plt.figure(figsize=(12,6))
sns.lineplot(data=df[df['category']=='A'], x='date', y='total_sales_ma', label='Category A (7-day MA)')
sns.lineplot(data=df[df['category']=='B'], x='date', y='total_sales_ma', label='Category B (7-day MA)')
plt.xticks(rotation=45)
plt.title('7-Day Moving Average of Total Sales')
plt.legend()
plt.show()
```

```python
#Case 2 - Customer Spending by Segment:
import pandas as pd

df = pd.read_csv(r"C:\Users\Dell\Desktop\HCAI5DS02_SauravKafle\venv\customer_segments.csv")
print(df.columns)

#– Create a violin plot and strip plot of transaction amount for each segment. What can you infer about skewness, modality, and outliers?
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.violinplot(x='segment', y='transaction_amount', data=df, inner=None, palette='Pastel1')
sns.stripplot(x='segment', y='transaction_amount', data=df, color='black', size=3, jitter=0.2)
plt.title('Violin & Strip Plot of Transaction Amount by Segment')
plt.show()

#- Plot a density plot (KDE) of transaction amount per segment. Which segment showsmore variability? Which one has higher spending tendencies?
plt.figure(figsize=(10,6))
sns.kdeplot(data=df[df['segment']=='Student']['transaction_amount'], fill=True, label='Student', alpha=0.5)
sns.kdeplot(data=df[df['segment']=='Working Professionals']['transaction_amount'], fill=True, label='Working Professionals', alpha=0.5)
plt.title('Density Plot of Transaction Amount by Segment')
plt.legend()
plt.show()


```

```python
# – Compute mean, median, IQR, and CV for each group and display them in a summary table.
import numpy as np

summary_stats = []

for segment in df['segment'].unique():
    data = df[df['segment'] == segment]['transaction_amount']
    mean = data.mean()
    median = data.median()
    q75, q25 = np.percentile(data, [75 ,25])
    iqr = q75 - q25
    cv = (data.std() / mean) * 100
    summary_stats.append([segment, mean, median, iqr, cv])

summary_df = pd.DataFrame(summary_stats, columns=['Segment', 'Mean', 'Median', 'IQR', 'CV (%)'])
print(summary_df)
```

```python
# # – Construct a boxplot with swarm overlay to visualize typical vs. exceptional spending behavior. Do the visuals agree with summary statistics?
print(df.columns)

plt.figure(figsize=(10,6))
sns.boxplot(x='segment', y='transaction_amount', data=df, palette='Set2')
sns.swarmplot(x='segment', y='transaction_amount', data=df, color='black', size=3)
plt.title('Boxplot with Swarm of Transaction Amount by Segment')
plt.show()



```
